**Data Visualization**

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

**Team Performance Analysis**

01. Track Changes in Team Rankings Over Time

```python
sns.set(style='darkgrid')

team_wins_per_season = match_data.groupby(['season', 'winner']).size().reset_index(name='wins')
team_wins_pivot = team_wins_per_season.pivot(index='winner', columns='season', values='wins').fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(team_wins_pivot, annot=True, cmap="viridis", linewidths=.5)
plt.title('Team Wins per Season')
plt.xlabel('Team')
plt.ylabel('Season')

plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

02. Analyze Winning Streaks and Performance Against Opponents

```python
match_data['team_win'] = match_data['winner'] == match_data['team1']
streaks = match_data.groupby('team1')['team_win'].apply(lambda x: x.cumsum() - x.cumsum().where(~x).ffill().fillna(0).astype(int)).reset_index()

head_to_head = match_data.groupby(['team1', 'team2', 'winner']).size().reset_index(name='wins')
head_to_head_pivot = head_to_head.pivot_table(index='team1', columns='team2', values='wins', aggfunc='sum').fillna(0)

plt.figure(figsize=(14, 8))
sns.heatmap(head_to_head_pivot, cmap='coolwarm', linewidths=.5, annot=True)
plt.title('Head-to-Head Performance')
plt.xlabel('Opponent Team')
plt.ylabel('Team')

plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

**Player Performance Analysis**

01. Identify Top Performers and Assess Player Consistency

```python
top_batters = deliveries_data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_bowlers = deliveries_data.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)


plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
top_batters.plot(kind='bar', color='skyblue')
plt.title('Top 10 Batters')
plt.ylabel('Total Runs')

plt.subplot(1, 2, 2)
top_bowlers.plot(kind='bar', color='salmon')
plt.title('Top 10 Bowlers')
plt.ylabel('Total Wickets')

plt.tight_layout()


plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

02. Evaluate the Impact of Different Player Roles on Match Outcomes

```python
role_performance = deliveries_data.groupby(['match_id', 'batter', 'bowler']).agg({
    'batsman_runs': 'sum',
    'is_wicket': 'sum'
}).reset_index()

merged_data = pd.merge(role_performance, match_data, left_on='match_id', right_on='id')

# Plot the impact of roles on match outcomes
sns.lmplot(x='batsman_runs', y='is_wicket', hue='result', data=merged_data, aspect=1.5)
plt.title('Impact of Player Roles on Match Outcomes')


plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

03. Top 10 Players of the Match

```python
top_players = match_data['player_of_match'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_players, y=top_players.index, palette='deep')
plt.title('Top 10 Players of the Match')
plt.xlabel('Number of Awards')
plt.ylabel('Player')

plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

**Match Outcome Analysis**

01. Effectiveness of the toss choice

```python
toss_effectiveness = match_data.groupby(['toss_decision', 'winner']).size().reset_index(name='wins')
toss_effectiveness_pivot = toss_effectiveness.pivot(index='toss_decision', columns='winner', values='wins').fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(toss_effectiveness_pivot, cmap='viridis', linewidths=.5, annot=True)
plt.title('Effectiveness of Toss Choice')
plt.xlabel('Winning Team')
plt.ylabel('Toss Decision')


plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```

02. Analyze Characteristics of High-Scoring Matches

```python
match_data['result_margin'] = pd.to_numeric(match_data['result_margin'], errors = 'coerce')
high_scoring_matches = match_data[match_data['result_margin'] > 100]

plt.figure(figsize=(12, 6))
sns.countplot(y=high_scoring_matches['venue'])
plt.title('Venues for High-Scoring Matches')

plt.savefig("C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png")
plt.show()

from IPython.display import FileLink
display(FileLink('C:/Data_Analytics/EDA_IPL_ANALYSIS_Project/Plots.png'))
```
